```python
img_mobil = cv2.imread("gambar-mobil.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_mobil.shape)

img_boneka = cv2.imread("gambar-boneka.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_boneka.shape)
```

```python
#import library untuk plot nya
import numpy as np
import matplotlib.pyplot as plt

def filter_batas(image):
  buf = image.copy()
  for h in range(1, image.shape[0]):
    for w in range(1, image.shape[1]):
      min_pixel = np.min(image[h-1:h+1, w-1:w+1])
      max_pixel = np.min(image[h-1:h+1, w-1:w+1])

      if image[h, w] < min_pixel:
        buf[h, w] = min_pixel
      elif image[h, w] > min_pixel:
        buf[h, w] = max_pixel
      else:
        buf[h, w] = image[h, w]

  return buf

img_batas_mobil = filter_batas(img_mobil)
img_batas_boneka = filter_batas(img_boneka)
```

```python


fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].imshow(img_mobil, cmap = "gray")
axes[0, 1].imshow(img_batas_mobil, cmap = "gray")

axes[1, 0].imshow(img_boneka, cmap = "gray")
axes[1, 1].imshow(img_batas_boneka, cmap = "gray")

plt.show()
```

```python
img_mobil = cv2.imread("gambar-mobil.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_mobil.shape)

img_boneka = cv2.imread("gambar-boneka.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_boneka.shape)
```

```python
def filter_rerata(image):
  buf = image.copy()
  for h in range(1, image.shape[0]):
    for w in range(1, image.shape[1]):
      buf[h, w] = np.mean(image[h-1:h+1, w-1:w+1])


  return buf

img_rerata_mobil = filter_rerata(img_mobil)
img_rerata_boneka = filter_rerata(img_boneka)
```

```python


fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].imshow(img_mobil, cmap = "gray")
axes[0, 1].imshow(img_rerata_mobil, cmap = "gray")

axes[1, 0].imshow(img_boneka, cmap = "gray")
axes[1, 1].imshow(img_rerata_boneka, cmap = "gray")

plt.show()
```

```python
img_mobil = cv2.imread("gambar-mobil.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_mobil.shape)

img_boneka = cv2.imread("gambar-boneka.jpg" , cv2.IMREAD_GRAYSCALE)  
print(img_boneka.shape)
```

```python
def filter_median(image):
  buf = image.copy()
  for h in range(1, image.shape[0]):
    for w in range(1, image.shape[1]):
      buf[h, w] = np.median(image[h-1:h+1, w-1:w+1])


  return buf

img_median_mobil = filter_median(img_mobil)
img_median_boneka = filter_median(img_boneka)
```

```python
fig, axes = plt.subplots(2, 2, figsize=(10, 6))

axes[0, 0].imshow(img_mobil, cmap = "gray")
axes[0, 1].imshow(img_median_mobil, cmap = "gray")

axes[1, 0].imshow(img_boneka, cmap = "gray")
axes[1, 1].imshow(img_median_boneka, cmap = "gray")

plt.show()
```
